```python
# Importiamo le librerie necessarie
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Carichiamo il set di dati
df = pd.read_csv("zoo.csv")

# Separiamo la variabile di output dalle variabili predittive
X = df.drop(['animal_name', 'class_type'], axis=1)

# Normalizziamo i dati
X_norm = (X - X.mean()) / X.std()

# Selezioniamo il numero ottimale di cluster
scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_norm)
    scores.append(silhouette_score(X_norm, kmeans.labels_))

# Visualizziamo il grafico del coefficiente di silhouette
import matplotlib.pyplot as plt
plt.plot(range(2, 11), scores)
plt.xlabel('Numero di cluster')
plt.ylabel('Coefficiente di silhouette')
plt.show()

# Addestriamo un modello K-Means con il numero ottimale di cluster
n_clusters = np.argmax(scores) + 2
kmeans_model = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_model.fit(X_norm)

# Aggiungiamo le etichette dei cluster al set di dati originale
df['cluster'] = kmeans_model.labels_

# Visualizziamo i risultati del clustering
for i in range(n_clusters):
    cluster_df = df[df['cluster'] == i]
    print(f"Cluster {i+1}: {len(cluster_df)} animali")
    print(cluster_df['class_type'].value_counts())
    print(cluster_df.mean())
    print("\n")
```
